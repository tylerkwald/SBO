```python
index = 214
axialProfile = pd.DataFrame(columns = ['TempG','TempF', 'Void', 'SatTemp','Node'])
tempG = []
tempF = []
void = []
satTemp = []
for i in np.arange(2,20):
    tempG.append(df['temp_g_' + str(i)][index])
    tempF.append(df['temp_f_' + str(i)][index])
    void.append(df['void_f_' + str(i)][index])
    satTemp.append(df['sat_temp_' + str(i)][index])
axialProfile['TempF'] = tempF
axialProfile['TempG'] = tempG
axialProfile['Void'] = void
axialProfile['SatTemp'] = satTemp
axialProfile['Node'] = np.arange(2,20)
```

```python
#Boundary Conditions Plot
fig, ax1 = plt.subplots()

plt.clf
plt.xlabel('Node')
temps = ['TempF', 'TempG']
ax1.plot(axialProfile['Node'], axialProfile[temps], label = ['Fluid T', 'Gas T'])
ax1.plot(axialProfile['Node'], axialProfile['SatTemp'], 'r--', label = 'Sat T', linewidth = .75)
ax1.set_ylabel('Temperature (K)')
ax2 = ax1.twinx()
ax2.plot(axialProfile['Node'], axialProfile['Void'],'k' ,label = 'Void')
ax2.set_ylabel('Fluid Void Fraction')
ax2.set_title('Axial Profile at 100s')
ax2.set_xlim(2,19)
fig.legend()
```
